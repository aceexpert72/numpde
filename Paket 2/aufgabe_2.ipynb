```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Aufgabe 2

```python
def compute_error(u,xh,uHat):
    N = len(xh)
    M = 100*(N-1)+1
    xf = np.linspace(xh[0],xh[-1],M)
    yf = u(xf)
    yh_interp = np.interp(xf, xh, uHat)
    abs_diff = np.abs(yf - yh_interp)
    max_error = np.max(abs_diff)

    return max_error
```

# Aufgabe 3

```python
#Aufgabe 3
u = np.sin
k = np.array(range(1, 10))  
N = 2 ** k
a, b = 0, np.pi

max_errors = []
hk = []

for n in N:
    xh = np.linspace(a, b, n+1)
    yh = u(xh)
    max_error = np.round(compute_error(u,xh,yh),3)
    hk_i = np.round(np.pi/(n-1),3)
    hk.append(hk_i)
    max_errors.append(max_error)



plt.figure(figsize=(8, 5))
plt.bar(np.arange(1, len(hk)+ 1), max_errors, width=0.6)
plt.xlabel('lenght hk')
plt.ylabel('Max Errors')
plt.title('Max Errors/hk')  
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(8, 5))
plt.loglog(hk,max_errors, marker='o', linestyle='dotted')
plt.xlabel('hk')
plt.ylabel('Max Errors')
plt.title('Log-log Max Errors/hk')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  
plt.show()
```
